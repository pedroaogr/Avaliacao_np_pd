```python
#Atribui a variável url o endereço do site onde está o banco de dados Copas_do_Mundo.csv no github
url = "https://raw.githubusercontent.com/pedroaogr/Avaliacao_np_pd/main/Copas_do_Mundo.csv"
#A parte df = pd.read_csv(url) lê o banco de dados que está no endereço do github acima e transforma em uma tabela chamada df
#A parte sep=';' define o ponto e vírgula como caractere separador para as colunas e
#o encoding='latin-1' escolhe o padrão latino para ler os caracteres com acentos, ç, etc fazendo uma correção sem quebrar a saída
df = pd.read_csv(url, sep=';', encoding='latin-1')
```

```python
#Mostrar um resumo do banco de dados
df.info()
```

```python
#Mostrar apenas o nome das colunas da tabela
df.columns
```

```python
#Mostrar o número de linhas e colunas da tabela
df.shape
```

```python
#Mostra a soma dos dados nulos por coluna
df.isnull().sum()
```

```python
#Apagar a coluna 'Unnamed:13' que apresenta apenas dados nulos, atribui e a df a nova tabela sem dados nulos
#Mostrando as 5 primeiras linhas novamente
df = df.drop('Unnamed: 13', axis=1)
df.head()
```

```python
#Quesito_1: Quem ganhou mais vezes a Copa do Mundo?
#onde listamos todos os países que venceram a copa, começando pelo Brasil com 5 copas em ordem decrescente
df['Winner'].value_counts()
```

```python
#Quesito_2: Quais paises mais chegaram mais em finais?
#onde listamos a soma das vezes em que os países chegaram em finais vitórias mais vices em ordem decrescente,
#começando pelo Brasil como resultado
Finais = (df['Winner'].value_counts().add(df['Runners-Up'].value_counts(), fill_value=0).sort_values(ascending=False))
Finais
#Interessante o comando .add() faz a soma alinhando pelos nomes dos países e
#o comando fill_value=0 vai considerar como zero se o país não existir em uma das listas
#e .sort_values(ascending=False) ordena do maior para o menor
```

```python
#Quesito_3: Mostrar a média Geral dos gols marcados por Copas do Mundo:
df['Goals Scored'].mean().round(2)
#esse comando seleciona a coluna de gols totais marcados em cada Copa e nos mostra a média geral por copa
```

```python
#Quesito_4: Mostra a média Geral de gols por jogo
(df['Goals Scored'] / df['Matches Played']).mean().round(2)
#Onde mostramos média de gols por jogo considerando todas as Copas
```

```python
#Quesito_5: Mostra a média de gols por partida por Ano da Copa
#Onde criamos uma nova coluna para apresentar a média de gols por jogo de cada ano da copa
df['Media_de_Gols'] = (df['Goals Scored'] / df['Matches Played']).round(2)
df[['Year', 'Media_de_Gols']]
```

```python
#Quesito_6: Mostrar o número de copas aconteceram, a média de Gols marcados pelo número de jogos, e demais marcadores estatísticos gerais
#como desvio padrão, mediana, quartis, média mínima, média máxima numa única tabela de saída
df['Media_de_Gols'].describe().round(2)
```

```python
#Quesito_7: Quais paises mais sediaram a Copa
#onde identificamos que Itália, Alemanha, França, Brasil e Mexico estão empatados com duas copas cada
df['Country'].value_counts()
```

```python
#Quesito_8: Localizar qual a copa teve maior média de gols e mostrar as características gerais dessa copa
#onde localizamos a copa de 1954 na Suíça com a maior média de Gols
#mostrando o campeão a Alemanha, o vice a Hungria, terceiro e quartos lugares, gols marcados, etc
df.loc[df['Media_de_Gols'].idxmax()]
```

```python
#Quesito_9: Definir copas antes de 1990 como Antigas e as copas após 1990 como Modernas
#em seguida mostre as médias de gols marcados nas Copas Antigas e nas Modernas
#onde conseguimos identificar que antigamente tínhamos uma média de gols maior do que atualmente.
df['Epocas'] = df['Year'].apply(lambda x: 'Antigas' if x < 1990 else 'Modernas')
df.groupby('Epocas')['Media_de_Gols'].mean().round(2)
```

```python
#Quesito_10: Mostre no gráfico a evidência encontrada no quesito 9
#ou seja que a evolução média de gols por copa diminuiu nas copas mais modernas
#podemos ver que essa curva diminui especialmente a partir da década de 1960
plt.plot(df['Year'], df['Media_de_Gols'])
plt.xlabel('Ano')
plt.ylabel('Média de Gols')
plt.title('Evolução da Média de Gols por Copa')
plt.show()
```
