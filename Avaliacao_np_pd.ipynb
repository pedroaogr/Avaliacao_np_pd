```python
url = "https://raw.githubusercontent.com/pedroaogr/Avaliacao_np_pd/main/Copas_do_Mundo.csv"

df = pd.read_csv(url, sep=';', encoding='latin-1')

df.head()
```

```python
df.info()
df.columns
df.shape
```

```python
#Quesito_1: Quem ganhou mais vezes a Copa do Mundo?
#onde listamos todos os países que venceram a copa, começando pelo Brasil com 5 copas em ordem decrescente
df['Winner'].value_counts()
```

```python
#Quesito_2: Quais paises mais chegaram em finais?
df['Winner'].value_counts() + df['Runners-Up'].value_counts()
```

```python
#Quesito_3: Mostrar a média Geral dos gols marcados em todas as Copas do Mundo:
df['Goals Scored'].mean().round(2)
```

```python
#Quesito_4: Mostra a média de gols por partida por Ano da Copa
df['Media_de_Gols'] = df['Goals Scored'] / df['Matches Played']
df[['Year', 'Media_de_Gols']]
```

```python
#Quesito_5: Mostrar o número de copas, a média de Gols marcados pelo número de copas, e demais marcadores estatísticos gerais
#como desvio padrão, mediana, quartis, média mínima, média máxima
df['Media_de_Gols'].describe().round(2)
```

```python
#Quesito_6: Quais paises mais sediaram a Copa
#onde identificamos que Itália, Alemanha, França, Brasil e Mexico estão empatados
df['Country'].value_counts()
```

```python
#Quesito_7: Localizar qual a copa teve maior média de gols e mostrar as características gerais dessa copa
#onde localizamos a copa de 1954 na Suíça com a maior média de Gols
df.loc[df['Media_de_Gols'].idxmax()]
```

```python
#Quesito_8: Definir copas antes de 1990 como Antigas e as copas após 1990 como Modernas
#em seguida mostre as médias de gols marcados nas Copas Antigas e nas Modernas
#onde conseguimos identificar que antigamente tínhamos uma média de gols maior do que atualmente.
df['Epocas'] = df['Year'].apply(lambda x: 'Antigas' if x < 1990 else 'Modernas')
df.groupby('Epocas')['Media_de_Gols'].mean().round(2)
```
